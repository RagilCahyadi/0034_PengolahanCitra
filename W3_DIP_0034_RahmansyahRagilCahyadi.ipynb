```python
import cv2
import matplotlib.pyplot as plt
import numpy as numpy
```

```python
# Baca Gambar
img = cv2.imread('dft.tif', cv2.IMREAD_GRAYSCALE)

# Brightness adjustment
brightness = 180
result = img.astype('int16') + brightness
result = result.clip(0, 255).astype('uint8')

# Tampilkan
plt.subplots(figsize=(10, 10))

plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title('Brightness +180')
plt.imshow(result, cmap='gray')
plt.axis('off')

plt.show()
```
