```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Baca Gambar dari file
image_bgr = cv2.imread('apel.jpg')

# Resize agar tidak terlalu besar
image_bgr = cv2.resize(image_bgr, (300, 200))

# Cek tipe data: ternyata sudah array NumPy
print("Tipe data:", type(image_bgr))
print("Shape:", image_bgr.shape)
print("Dtype:", image_bgr.dtype)
print()

print("Artinya:")
print(f" Tinggi = {image_bgr.shape[0]} piksel")
print(f" Lebar = {image_bgr.shape[1]} piksel")
print(f" Channel = {image_bgr.shape[2]} (BGR)")

# Konversi BGR -> RGB untuk ditampilkan di matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

print("Total piksel:", image_bgr.shape[0] * image_bgr.shape[1])

#Tampilkan gambar
plt.figure(figsize=(4, 6))
plt.imshow(image_rgb)
plt.title("Gambar Apel\n(sudah dikonversi BGR -> RGB)")
plt.axis('off')
plt.show()
```

```python
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

# Ambil area kecil 8x8 piksel dari tengah gambar (area apel)
area_kecil = image_gray[192:200, 100:108]

print("=== MATRIKS f(x,y) - Area 8x8 Piksel ===")
print("Setiap angka = nilai intensitas 1 piksel (0=hitam, 255= putih)")
print()
print(area_kecil)
print()
print("Contoh baca:", "f(100, 199) =", image_gray[199, 100])
print("Contoh baca:", "f(101, 198) =", image_gray[198, 101])

# Tampilkan area kecil dengan nilai angkanya
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Kiri: gambar grayscale penuh
axes[0].imshow(image_gray, cmap="gray", vmin=0, vmax=255)
axes[0].set_title("Gambar Grayscale\n(setiap piksel = 1 angka)")
axes[0].axis("off")

# Tandai area yang kita zoom
import matplotlib.patches as patches
rect = patches.Rectangle((150, 100), 8, 8, linewidth=2, edgecolor='red', facecolor='none')
axes[0].add_patch(rect)
axes[0].text(150, 95, 'area ini\ndi zoom', color='red', fontsize=8)

# Kanan: tampilkan nilai angka pada setiap piksel
axes[1].imshow(area_kecil, cmap='gray', vmin=0, vmax=255)
for i in range(8):
  for j in range(8):
    nilai = area_kecil[i, j]
    warna_teks = 'white' if nilai < 128 else 'black'
    axes[1].text(j, i, str(nilai), ha='center', va='center', fontsize=9,
                 color = warna_teks, fontweight='bold')

axes[1].set_title("Zoom 8x8 = nilai tiap piksel")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
bgr_area_kecil = image_bgr[100:108, 175:183]

print("Nilai MATRIKS")
print()
print(bgr_area_kecil)
print()


rgb_area_kecil = cv2.cvtColor(bgr_area_kecil, cv2.COLOR_BGR2RGB)

total_pixel = image_bgr.shape[0] * image_bgr.shape[1]
ukuran_bytes = image_bgr.nbytes
ukuran_mb = ukuran_bytes / (1024 ** 2)

fig_2, axes_2 = plt.subplots(1, 2, figsize=(16, 8))

axes_2[0].imshow(image_rgb)
axes_2[0].set_title('Gambar RGB\n(setiap piksel = 3 angka)')
axes_2[0].axis('off')

rect = patches.Rectangle((175, 100), 8, 8, linewidth=2, edgecolor='blue', facecolor='none')
axes_2[0].add_patch(rect)
axes_2[0].text(175, 95, 'area zoom', color='blue', fontsize=10, fontweight='bold')


axes_2[1].imshow(rgb_area_kecil)
for i in range(8):
  for j in range(8):
    nilai_bgr = bgr_area_kecil[i, j]
    offsets = [-0.2, 0, 0.2]
    colors = ['blue', 'green', 'red']
    for mn in range(3):
      nilai_satuan = nilai_bgr[mn]
      axes_2[1].text(j, i + offsets[mn], str(nilai_satuan),
                     ha='center', va='center', fontsize=8,
                     color=colors[mn], fontweight='bold')

axes_2[1].set_title('Zoom 8x8 piksel\nNilai [B, G, R]')
axes_2[1].axis('off')
plt.suptitle('Representasi RGB (3 Channels)', fontweight='bold')
plt.tight_layout()
plt.show()

print()
print("Shape:", image_bgr.shape)
print("Total Pixel:", total_pixel)
print("Ukuran bytes:", ukuran_bytes)
print("Ukuran MB:", f"{ukuran_mb:.2f} MB")
```

### Penjelasan Kode Perbaris

Berikut adalah penjelasan teknis mengenai fungsi dari baris-baris kode di atas:

1.  **`bgr_area_kecil = image_bgr[100:108, 175:183]`**: Melakukan *slicing* pada array NumPy `image_bgr` untuk mengambil sub-matriks berukuran 8x8 piksel, mulai dari baris 100-107 dan kolom 175-182.
2.  **`rgb_area_kecil = cv2.cvtColor(bgr_area_kecil, cv2.COLOR_BGR2RGB)`**: Mengonversi format warna area kecil tersebut dari BGR (standar OpenCV) ke RGB agar warna tampil akurat saat divisualisasikan dengan Matplotlib.
3.  **`total_pixel = image_bgr.shape[0] * image_bgr.shape[1]`**: Menghitung total piksel gambar dengan mengalikan dimensi tinggi (index 0) dan lebar (index 1).
4.  **`ukuran_bytes = image_bgr.nbytes`**: Mengambil total ukuran memori yang digunakan oleh array gambar dalam satuan bytes.
5.  **`ukuran_mb = ukuran_bytes / (1024 ** 2)`**: Mengonversi ukuran dari bytes ke Megabytes (MB) dengan membaginya sebanyak dua kali dengan 1024.
6.  **`fig_2, axes_2 = plt.subplots(1, 2, figsize=(16, 8))`**: Membuat *figure* baru yang berisi dua buah sub-plot berdampingan (1 baris, 2 kolom) dengan ukuran kanvas 16x8 inci.
7.  **`axes_2[0].imshow(image_rgb)`**: Menampilkan gambar asli dalam format RGB pada sub-plot pertama (kiri).
8.  **`axes_2[0].set_title(...)` & `axes_2[0].axis('off')`**: Memberikan judul pada sub-plot pertama dan menyembunyikan garis sumbu (axis).
9.  **`rect = patches.Rectangle((175, 100), 8, 8, ...)`**: Mendefinisikan objek kotak (patch) untuk menandai lokasi koordinat (175, 100) dengan lebar dan tinggi 8 piksel.
10. **`axes_2[0].add_patch(rect)`**: Menambahkan kotak biru tersebut ke dalam sub-plot gambar asli sebagai penanda visual area yang di-*zoom*.
11. **`axes_2[1].imshow(rgb_area_kecil)`**: Menampilkan potongan gambar 8x8 piksel yang sudah di-*zoom* pada sub-plot kedua (kanan).
12. **`for i in range(8): for j in range(8):`**: Melakukan perulangan melalui setiap baris (i) dan kolom (j) dari matriks 8x8 tersebut.
13. **`nilai_bgr = bgr_area_kecil[i, j]`**: Mengambil nilai warna [Blue, Green, Red] dari satu piksel spesifik pada posisi (i, j).
14. **`for mn in range(3):`**: Perulangan untuk mengakses masing-masing dari 3 channel warna (0: Biru, 1: Hijau, 2: Merah).
15. **`axes_2[1].text(...)`**: Menuliskan angka intensitas warna secara manual di atas gambar *zoom*. Posisi teks digeser sedikit menggunakan `offsets` agar angka B, G, dan R tidak tumpang tindih.
16. **`plt.tight_layout()` & `plt.show()`**: Mengatur tata letak agar tidak ada teks yang terpotong dan menampilkan hasil visualisasi ke layar.
